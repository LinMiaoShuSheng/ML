# 自己练习

```python
import numpy as np
import pandas as pd
```

```python
# y = x1 + x2
df = pd.DataFrame({'x1':[1,2,3,4,5,6],'x2':[1,2,1,2,1,2],'y':[2,4,4,6,6,8]})
df
```

```python
x = df[['x1','x2']]
x = x.values.reshape(-1,2)
y = df['y']
y = y.values.reshape(-1,1)
print(y.shape)
theta = np.mat(x.T.dot(x)).I.dot(x.T).dot(y)
print(theta)
```

```python
#批量梯度下降算法 BGD 
#初始化theta和alpha
theta = np.array([[0],[0]])
alpha = 0.01
```

```python
the_end = 1e-6
max_iter = 500
for time in range(max_iter):
    theta = theta - alpha*x.T.dot(x.dot(theta)-y)
    loss = np.power(x.dot(theta)-y,2).sum()/2
    print("loss值为%f\n theta值为\n"%loss,theta)
    if loss < the_end:
        break
```

```python
#随机梯度下降算法 
#初始化theta，alpha
theta_s = np.array([[0],[0]])
alpha_s = 0.01
```

```python
#这里其实也需要和上一个一样，设置一个max_iter，不断重复直至收敛，即外层还需套一层循环，光遍历完一遍所有的还不够
for i in range(6):
    theta_s = theta_s-alpha_s * (x[i].dot(theta_s)-y[i])*x[i]
    loss = np.power(x[i].dot(theta_s)-y[i],2).sum()/2
    print("loss:%f\n theta:"%loss,theta_s)
```

```python
#小批量梯度下降
#初始化theta和alpha
theta_m = np.array([[0],[0]])
alpha_m = 0.01
```

```python
#这里其实也需要和上一个一样，设置一个max_iter，不断重复直至收敛，即外层还需套一层循环，光遍历完一遍所有的还不够
for i in range(3):
    theta_m = theta_m - alpha_m * x[2*i:2*i+2].dot(x[2*i:2*i+2].dot(theta_m)-y[2*i:2*i+2])
    loss = np.power(x[2*i:2*i+2].dot(theta_m)-y[2*i:2*i+2],2).sum()/2
    print("loss:%f\n theta:"%loss,theta_m)
```
